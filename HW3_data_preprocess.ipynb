```python
import pandas as pd
import numpy as np
data = pd.read_excel("/Users/guoyixuan/Documents/pythoncode/ccwapp/2023 QS World University Rankings.xlsx")
data.head()
```

```python
#data_name = data.to_numpy() # dataframe to numpy
list = data[2:3].values.tolist() # numpy to list
list_split = []
for i in range(data.shape[1]):
    list_split.append(list[0][i].split("\n"))
```

```python
strin = []
for i in range(data.shape[1]):
    strin.append(''.join(list_split[i]))
```

```python
data.columns = strin
data.head()
```

```python
data.drop(data.index[0:3], inplace=True)
data.head()
```

```python
#data_us = data.loc[data.location == "United States"]
#data_us.head()
data["location"] == "United States"
```

```python
# 找出美國大學
# 201
data_us = data[data["location"] == "United States"]
data_us.head()
```

```python
# 找出會用到的欄位
dropcol = ["rank display", "rank display2", "location", "size", "focus", "research", "age band", "status"]
data_choose = data_us.drop(dropcol, axis=1)
#data_choose = data_us.loc[:, ["rank display2", "location", "size", "focus", "research", "age band", "status"]]
data_choose.index = data_choose.index - 2
print(data_choose.index)
```

```python
data_ranking = data_choose.index.to_numpy()
data_choose.insert(loc = 0, column = "ranking", value = data_ranking)
```

```python
data_choose.fillna(0)
data_choose.replace(np.nan, 0, inplace=True)
```

```python
data_choose.to_csv("/Users/guoyixuan/Documents/pythoncode/ccwapp/2023 QS World University Rankings_US.csv", index=False, encoding="utf-8-sig")
```
