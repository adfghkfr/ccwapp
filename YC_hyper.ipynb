```python
from PyQt6 import QtWidgets, uic
import pyqtgraph as pg
import numpy as np
from scipy.stats import norm
from scipy.stats import t
import sys
import pyqtgraph as pg
 


class MainWindow(QtWidgets.QMainWindow):
 
    def __init__(self, *args, **kwargs):
        super(MainWindow, self).__init__(*args, **kwargs)
 
        #Load the UI Page by PyQt6
        uic.loadUi('hypothesis.ui', self)
        self.setWindowTitle('PyQtGraph shows normal distribution')
       
        # Signals
        #--------繪圖--------------------
        self.LE_mu_a.returnPressed.connect(self.update_plot)
        self.LE_sigma.returnPressed.connect(self.update_plot)
        self.LE_alpha.returnPressed.connect(self.update_plot)
        
        self.LE_sample.returnPressed.connect(self.update_plot)
        self.LE_beta.returnPressed.connect(self.update_plot)
        
        #---------計算 beta--------------
        self.LE_mu_a.returnPressed.connect(self.compute_beta)
        self.LE_sigma.returnPressed.connect(self.compute_beta)
        self.LE_alpha.returnPressed.connect(self.compute_beta)
        
        self.LE_sample.returnPressed.connect(self.compute_beta)
        self.LE_beta.returnPressed.connect(self.compute_beta)
        self.update_plot()
        
        
        
        self.LE_beta.returnPressed.connect(self.compute_sample)
    # Slots
    def update_plot(self):
       
        self.gViews.clear() # clear current plot before plotting
        
        # mu=int((self.LE_mu.text()))
        mu = 28
        mu_a = float((self.LE_mu_a.text()))
        sigma=float((self.LE_sigma.text()))
        sample =  int((self.LE_sample.text()))
        
        x = np.linspace(mu-4*sigma , mu+4*sigma, 1000)
        x_mu = np.linspace(mu_a - 4*sigma , mu_a +4*sigma , 1000)
        y = norm.pdf(x,loc=mu , scale=sigma)
        y_mu = norm.pdf(x_mu ,loc=mu_a , scale=sigma)
        
        # Calculate critical values for alpha area
        alpha = float((self.LE_alpha.text()))
        Z_alpha = norm.ppf(alpha/2)
        
        beta = float((self.LE_beta.text()))
        Z_beta = norm.ppf(beta/2)
        sample =  int((self.LE_sample.text()))
        upper_bound = mu - Z_alpha * sigma 
        lower_bound = mu + Z_alpha * sigma

        # Create lines for alpha area
        x_alpha_left = np.linspace(mu-4*sigma , lower_bound, 1000)
        y_alpha_left = norm.pdf(x_alpha_left , loc=mu , scale=sigma)
        
        
        
        cur1 = self.gViews.plot(x_alpha_left , y_alpha_left , pen = 'r', name = 'Demo')
        cur2 = self.gViews.plot(x_alpha_left, np.zeros(len(y_alpha_left)))
        
        x_alpha_right = np.linspace(upper_bound , mu+4*sigma , 1000)
        y_alpha_right = norm.pdf(x_alpha_right, loc=mu , scale=sigma)
        patchcur = pg.FillBetweenItem(curve1 = cur1 , curve2 = cur2, \
            brush = 'green')
        self.gViews.addItem(patchcur)
        
        
       
        cur3 = self.gViews.plot(x_alpha_right , y_alpha_right , pen = 'r', name = 'Demo')
        cur4 = self.gViews.plot(x_alpha_right, np.zeros(len(y_alpha_right)))
        patchcur1 = pg.FillBetweenItem(curve1 = cur3 , curve2 = cur4, \
            brush = 'green')
        self.gViews.addItem(patchcur1)
        
        
        upper_bound = mu - Z_beta * sigma 

        # Create lines for alpha area
        x_beta_right = np.linspace(upper_bound ,mu_a +4*sigma , 1000)
        y_beta_right = norm.pdf(x_beta_right , loc=mu_a , scale=sigma)
        
        
        
#         cur5 = self.gViews.plot(x_beta_right , y_beta_right , pen = 'r', name = 'Demo')
#         cur6 = self.gViews.plot(x_beta_right, np.zeros(len(y_beta_right)))
        
#         patchcur3 = pg.FillBetweenItem(curve1 = cur5 , curve2 = cur6, \
#             brush = 'y')
#         self.gViews.addItem(patchcur3)
        
        
            
        pen = pg.mkPen(color=(255, 0, 0), width = 5) # Qt.DotLine, Qt.DashDotLine and Qt.DashDotDotLine
        pen_1 = pg.mkPen(color='y', width = 5) # Qt.DotLine, Qt.DashDotLine and Qt.DashDotDotLine

        
        cur1 = self.gViews.plot(x, y, pen = pen, name = 'Demo')
        cur2 = self.gViews.plot(x_mu , y_mu , pen =  pen_1, name = 'Demo')
        self.gViews.setYRange(0, max(y)+0.01)
        
        
        
        
        # patchcur = pg.FillBetweenItem(curve1=cur1, curve2=cur3, brush='g')
        # self.gViews.addItem(patchcur)
        # self.vLine = pg.InfiniteLine(pos=x_value, angle=90, movable=False)
        # self.gViews.addItem(self.vLine)


    

     
    def compute_beta(self):
        mu = 28
        mu_a = int((self.LE_mu_a.text()))
        sigma = float((self.LE_sigma.text()))
        sample =  int((self.LE_sample.text()))
        alpha = float((self.LE_alpha.text()))
        Z_alpha = norm.ppf(alpha)
        critical_value = mu + Z_alpha * sigma / np.sqrt(sample)
        beta = (critical_value - mu_a) / (sigma / np.sqrt(sample))
        p_beta = norm.cdf( beta)
        self.LE_beta.setText(str(abs(p_beta-1)))
        
        self.update_plot()
        
        
    def compute_sample(self):
        mu = 28
        mu_a = int((self.LE_mu_a.text()))
        sigma = float((self.LE_sigma.text()))
        sample =  int((self.LE_sample.text()))
        alpha = float((self.LE_alpha.text()))
        beta=float((self.LE_beta.text()))
        Z_alpha = norm.ppf(alpha)
        
        n=((Z_alpha+beta)**2)*(sigma**2)/(mu_a-mu)**2
        self.LE_sample.setText(str(int(n)))
        self.LE_beta.setText(str(beta))
        self.update_plot()
     
#----------- Beta 估計 ---------------------
#         mu = 40
#         mu_a = 38.5
#         alpha = 0.05
#         n = 64
#         sigma = 5

#         # 計算 alpha 對應的 z 值
#         z_alpha = norm.ppf(alpha)
#         k= mu+ z_alpha*(sigma / np.sqrt(n))
#         # 計算 beta 值
#         z_beta = ((k-mu_a)/(sigma / np.sqrt(n)))
#         p_power = norm.cdf(z_beta)
#         print("Z_Beta =",z_beta)
#         print("Power =",p_power)
#         print("Beta =",1-p_power)
#---------------------------------------------------        
 
def main():
    app = QtWidgets.QApplication(sys.argv)
    main = MainWindow()
    main.show()
    sys.exit(app.exec())
 
if __name__ == '__main__':
    main()
```
