```python
#評估不能用的點的比例
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt
import random
#unifrom
size = 10**5
random.seed(10)
xx = np.random.uniform(low = -1.5, high = 1.5, size = size)
yy = np.random.uniform(low = -1, high = 1, size= size)
ind = (xx**2) * (1 + np.sin(3*yy) * np.cos(8*xx)) + (yy**2) * (2 + np.cos(5*xx) * np.cos(8*yy)) 
#ind = (xx**2) * (1 + np.sin(5*yy) * np.cos(1*xx)) + (yy**2) * (1 + np.cos(2*xx) * np.cos(2*yy)) 
plt.scatter(xx, yy, color = "black")
plt.scatter(xx[ind<1], yy[ind<1], color = "red")
#plt.xlim(-1, 2)
#plt.ylim(-2, 2)
plt.show()
print(sum(ind<1)/size) 

#normal
xx = np.random.normal(loc = 0, scale = 1, size = size)
yy = np.random.normal(loc = 2, scale = 5, size= size)
ind = (xx**2) * (1 + np.sin(3*yy) * np.cos(8*xx)) + (yy**2) * (2 + np.cos(5*xx) * np.cos(8*yy)) 
plt.scatter(xx, yy, color = "black")
plt.scatter(xx[ind<1], yy[ind<1], color = "red")
plt.show()
print(sum(ind<1)/size) 

#beta
xx = np.random.beta(a = 1, b = 1, size = size)
yy = np.random.beta(a = 4, b = 4, size= size)
ind = (xx**2) * (1 + np.sin(3*yy) * np.cos(8*xx)) + (yy**2) * (2 + np.cos(5*xx) * np.cos(8*yy)) 
plt.scatter(xx, yy, color = "black")
plt.scatter(xx[ind<1], yy[ind<1], color = "red")
plt.show()
print(sum(ind<1)/size) 


xx = np.random.beta(a = 1, b = 1, size = size)
yy = np.random.normal(loc = 2, scale = 5, size= size)
ind = (xx**2) * (1 + np.sin(3*yy) * np.cos(8*xx)) + (yy**2) * (2 + np.cos(5*xx) * np.cos(8*yy)) 
plt.scatter(xx, yy, color = "black")
plt.scatter(xx[ind<1], yy[ind<1], color = "red")
plt.show()
print(sum(ind<1)/size) 
```
